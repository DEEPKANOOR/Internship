```python
# Answer4).  To search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on
#www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be 
#scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, 
#“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the 
#details is missing then replace it by “- “. Save your results in a dataframe and CSV.
```

```python
driver.get('https://www.flipkart.com/')
```

```python
search4=driver.find_element_by_xpath("//input[@class='_3704LK']")
button4=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
```

```python
search4.clear()
inp= str(input('enter the smartphone model which you want to search for \t'))
```

```python
search4.send_keys(inp)
button4.click()
```

```python
href_tags4=[]
url4=[]
name_tags4=[]
name4=[]
brand4=[]
color4=[]
price_tags4=[]
price4=[]
more_tags4=[]
ram4=[]
rom4=[]
disp_tags4=[]
disp4=[]
cam_tags4=[]
pcam4=[]
scam4=[]
bat_tags4=[]
bat4=[]
```

```python
href_tags4=driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in href_tags4:
    url4.append(i.get_attribute('href'))
name_tags4=driver.find_elements_by_xpath("//div[@class='_4rR01T']")
for j in name_tags4:
    name4.append(j.text.split('(')[0])
    brand4.append(j.text.split('(')[0].split(' ')[0])
    color4.append(j.text.split('(')[1].split(',')[0])
price_tags4=driver.find_elements_by_xpath("//div[@class='_30jeq3 _1_WHN1']")
for k in price_tags4:
    if k.text=='':
        price4.append('-')
    else:
        price4.append(k.text)
more_tags4=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[1]")
for l in more_tags4:
    if (l.text).split('|')[0]=='':
        ram4.append('-')
    else:
        ram4.append((l.text).split('|')[0])
    if (l.text).split('|')[1]=='':
        rom4.append('-')
    else:
        rom4.append((l.text).split('|')[1])
disp_tags4=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[2]")
for m in disp_tags4:
    if m.text=='':
        disp4.append('-')
    else:
        disp4.append(m.text.split(')')[0]+')')
cam_tags4=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[3]")
for n in cam_tags4:
    if '|' in n.text:
        if n.text.split('|')[1]=='':
            scam4.append('-')
        else:
            scam4.append(n.text.split('|')[1])
        if n.text.split('|')[0]=='':
            pcam4.append('-')
        else:
            pcam4.append(n.text.split('|')[0])
    else:
        if n.text=='':
            pcam4.append('-')
        else:
            pcam4.append(n.text)
            scam4.append('-')
bat_tags4=driver.find_elements_by_xpath("//div[@class='fMghEO']/ul/li[4]")
for o in bat_tags4:
    if o.text=='':
        bat4.append('-')
    else:
        bat4.append(o.text)
```

```python
print(len(name4),len(url4),len(pcam4),len(scam4),len(disp4),len(price4),len(brand4),len(ram4),len(rom4),len(color4),len(bat4))
```

```python
df4=pd.DataFrame({'Smartphone Name':name4,'Brand Name':brand4,'Price':price4,'RAM':ram4,'ROM':rom4,'Colour':color4,'Battery Capacity':bat4,'Display Size':disp4,'Primary Camera':pcam4,'Secondary Camera':scam4,'Product Url':url4})
df4
```

```python
df4.to_csv('Answer4.csv')
```

```python
# Answer 8). to scrape the details for all billionaires from www.forbes.com. Details to be 
# scrapped: “Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.
```

```python
a8=driver.find_element_by_xpath('/html/body/div[1]/header/nav/div[3]/ul/li[1]/div[2]/ul/li[2]/a')
driver.get(a8.get_attribute('href'))
```

```python
rank8=[]
name8=[]
worth8=[]
age8=[]
cit8=[]
src8=[]
ind8=[]
rank_tags8=[]
name_tags8=[]
worth_tags8=[]
age_tags8=[]
cit_tags8=[]
src_tags8=[]
ind_tags8=[]
```

```python
# for speed reason, m running itfor only first 5 pages..
```

```python
for a in range(5):
    rank_tags8=driver.find_elements_by_xpath("//div[@class='rank']")
    name_tags8=driver.find_elements_by_xpath("//div[@class='personName']")
    worth_tags8=driver.find_elements_by_xpath("//div[@class='netWorth']/div[1]")
    age_tags8=driver.find_elements_by_xpath("//div[@class='age']/div")
    cit_tags8=driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")
    src_tags8=driver.find_elements_by_xpath("//div[@class='source']")
    ind_tags8=driver.find_elements_by_xpath("//div[@class='category']/div")
    for i in rank_tags8:
        rank8.append(i.text)
    for j in name_tags8:
        name8.append(j.text)
    for k in worth_tags8:
        worth8.append(k.text)
    for l in age_tags8:
        if l.text=='':
            age8.append('-')
        else:
            age8.append(l.text)
    for m in cit_tags8:
        cit8.append(m.text)
    for n in src_tags8:
        if n.text=='':
            src8.append('-')
        else:
            src8.append(n.text)
    for o in ind_tags8:
        if o.text=='':
            ind8.append('-')
        else:
            ind8.append(o.text)
    delay=15
    next_button=driver.find_element_by_xpath("//button[@class='pagination-btn pagination-btn--next ']")
    try:
        next_button.click()
    except NoSuchElementException:
        WebDriverWait(driver,delay).until(EC.presence_of_element_located((By.CLASS_NAME,'fs-icon fs-icon--arrow-right')))
        next_button=driver.find_element_by_xpath("//button[@class='pagination-btn pagination-btn--next ']") 
        next_button.click()
    except TimeOutException:
        print('No more page available')
```

```python
print(len(rank8),len(name8),len(worth8),len(age8),len(cit8),len(src8),len(ind8))
```

```python
df8=pd.DataFrame({'rank':rank8[0:100],'name':name8[0:100],'age':age8[0:100],'citizenship':cit8[0:100],'source':src8[0:100],'industry':ind8[0:100],'net worth':worth8[0:100]})
df8
```
